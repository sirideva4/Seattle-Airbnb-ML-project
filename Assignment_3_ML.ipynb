# SVM

```python
from sklearn import svm
```

```python
X_train_scaled = X_train
X_test_scaled = X_test
```

```python
for column in X_train_scaled.columns:
    X_train_scaled[column] = (X_train_scaled[column] - X_train_scaled[column].min()) / (X_train_scaled[column].max() - X_train_scaled[column].min()) 
X_train_scaled
```

```python
for column in X_test_scaled.columns:
    X_test_scaled[column] = (X_test_scaled[column] - X_test_scaled[column].min()) / (X_test_scaled[column].max() - X_test_scaled[column].min()) 
X_test_scaled
```

```python
clf = svm.SVC(kernel = 'linear')
clf.fit(X_train_scaled, y_train)
```

```python
predicted_svm = clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
svm_matrix = confusion_matrix(y_test, predicted_svm)
print("Linear kernel")
print(svm_matrix)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predicted_svm, normalize=True)
```

```python
clf_poly = svm.SVC(kernel = 'linear', C = 1.8)
clf_poly.fit(X_train_scaled, y_train)
predicted_svm_poly = clf_poly.predict(X_test_scaled)
svm_matrix_poly = confusion_matrix(y_test, predicted_svm_poly)
print("Linear kernel with C = 1.8")
print(svm_matrix_poly)
accuracy_score(y_test, predicted_svm_poly, normalize=True)
```

```python
clf_poly = svm.SVC(kernel = 'poly', degree = 2)
clf_poly.fit(X_train_scaled, y_train)
predicted_svm_poly = clf_poly.predict(X_test_scaled)
svm_matrix_poly = confusion_matrix(y_test, predicted_svm_poly)
print("Polynomial kernel with degree = 2")
print(svm_matrix_poly)
accuracy_score(y_test, predicted_svm_poly, normalize=True)
```

```python
clf_poly = svm.SVC(kernel = 'poly', degree = 2, C = 1.5)
clf_poly.fit(X_train_scaled, y_train)
predicted_svm_poly = clf_poly.predict(X_test_scaled)
svm_matrix_poly = confusion_matrix(y_test, predicted_svm_poly)
print("Polynomial kernel with degree = 2, C = 1.5")
print(svm_matrix_poly)
accuracy_score(y_test, predicted_svm_poly, normalize=True)
```
